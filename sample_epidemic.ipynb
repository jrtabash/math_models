```python
import math_models_epidemic as epidemic

%matplotlib inline
```

```python
# SIR Model
#   Transmit Rate: 3.5
#   Remove Rate: 0.5

sirModel = epidemic.SIRModel(transmitRate=3.5, removeRate=0.5)
t, sir = epidemic.solve(sirModel, 15, 150)
epidemic.plot(sirModel, t, sir)
```

```python
# SEIR Model
#   Transmit Rate: 3.5
#   Infect Rate: 1.0
#   Remove Rate: 0.5

seirModel = epidemic.SEIRModel(transmitRate=3.5, infectRate=1.0, removeRate=0.5)
t, seir = epidemic.solve(seirModel, 20, 200)
epidemic.plot(seirModel, t, seir)
```

```python
# SEIR Model with Infectivity During Exposed Period
#   Transmit Rate: 3.5
#   Reduce Exposed Infectivity Rate: 0.25
#   Infect Rate: 1.0
#   Remove Rate: 0.5

seir2Model = epidemic.SEIRModel(transmitRate=3.5, reducedEIRate=0.25, infectRate=1.0, removeRate=0.5)
t, seir = epidemic.solve(seir2Model, 20, 200)
epidemic.plot(seir2Model, t, seir)
```
